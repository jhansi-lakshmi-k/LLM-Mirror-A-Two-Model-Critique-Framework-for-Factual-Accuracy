**flan-t5-large is bboth acting as. crtique and the one which answers to questions from the user. the same both is evaluating its answers**

```python
!pip install -q transformers accelerate

from transformers import pipeline

# Use a better small model with stronger reasoning
model_name = "google/flan-t5-large"

# Load the text2text-generation pipeline
generator = pipeline("text2text-generation", model=model_name)

print("Better model loaded and ready!")
```

```python
def get_answer_and_critique(question):
    # Step 1: Get the answer
    prompt_answer = f"Answer the following question:\n{question}"
    answer = generator(prompt_answer)[0]['generated_text'].strip()

    # Step 2: Critique the answer
    critique_prompt = f"""You are a helpful assistant.

Question: {question}
Answer: {answer}

Now critique the above answer for honesty, factual accuracy, and hallucination. Was anything misleading or unsupported? give the answer properly in few sentences and whether the actual model is saying right or not and factual or hallucination"""
    critique = generator(critique_prompt)[0]['generated_text'].strip()

    # Output
    print("🔹 Question:", question)
    print("\n🤖 Answer:\n", answer)
    print("\n🔍 Critique:\n", critique)
```

```python
get_answer_and_critique("")
```

**model flan-t5-base  is the one answering to user questions and flan-t5-large is the critique and evaluating the performance of the above model**

```python
from transformers import pipeline

# Answering model (stronger one)
model_answer = pipeline("text2text-generation", model="google/flan-t5-base")

# Critiquing model (different one)
model_critic = pipeline("text2text-generation", model="google/flan-t5-large")

print(" Both models loaded!")
```

```python
def double_model_critique(question):
    # Step 1: Generate answer using the larger model
    prompt_answer = f"Answer the following question:\n{question}"
    answer = model_answer(prompt_answer)[0]['generated_text'].strip()

    # Step 2: Critique using a separate model
    critique_prompt = f"""You are a different AI assistant.

Question: {question}
Answer: {answer}

Please critique the above answer for factual correctness, hallucination, or misleading claims. """

    critique = model_critic(critique_prompt)[0]['generated_text'].strip()

    # Output
    print("🔹 Question:", question)
    print("\n Answer (from flan-t5-base):\n", answer)
    print("\n Critique (from flan-t5-large):\n", critique)
```

```python
double_model_critique("where is Lubbock ?")
```
